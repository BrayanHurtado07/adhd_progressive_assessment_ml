```python
#  notebook/01_eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv("../data/snap_dataset_labeled.csv")

# Ver las primeras filas
df.head()

# Ver columnas
print("Columnas:", df.columns.tolist())

# Distribucin de perfiles
plt.figure(figsize=(6, 4))
sns.countplot(x='profile', data=df)
plt.title("Distribucin de perfiles cognitivos")
plt.show()

# Distribucin de severidades
plt.figure(figsize=(6, 4))
sns.countplot(x='severity', data=df)
plt.title("Distribucin de niveles de severidad")
plt.show()

# Estadsticas descriptivas por pregunta
df.iloc[:, 1:27].describe()

# Matriz de correlacin
plt.figure(figsize=(12, 10))
sns.heatmap(df.iloc[:, 1:27].corr(), annot=False, cmap="coolwarm")
plt.title("Correlacin entre respuestas SNAP-IV")
plt.show()

# Boxplot de algunas preguntas vs. perfil
plt.figure(figsize=(10, 4))
sns.boxplot(x='profile', y='1', data=df)
plt.title("Distribucin de respuestas a la pregunta 1 por perfil")
plt.show()

sns.boxplot(x='profile', y='10', data=df)
plt.title("Distribucin de respuestas a la pregunta 10 por perfil")
plt.show()
```
